```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df1 = pd.read_csv("tmdb_5000_credits.csv")
df2 = pd.read_csv("tmdb_5000_movies.csv")
```

#### Rename the column 'movie_id' of df1 to 'id'

```python
df1.rename(columns={'movie_id':'id'}, inplace=True)
```

#### Merge df2 with df1

```python
df2 = df2.merge(df1)
```

#### Sort df2 by 'popularity' and save it to a new dataframe 'pop'

```python
pop = df2.sort_values(['popularity'], ascending=False)
```

#### A Horizontal Bar chart of top10 movies by their popularity

```python
movie_title = pop['title'].head(10)
movie_popularity = pop['popularity'].head(10)
```

```python
%matplotlib inline
plt.barh(movie_title, movie_popularity)
plt.gca().invert_yaxis()
plt.xlabel('Popularity')
plt.title('Popular Movies')
plt.show()
```
